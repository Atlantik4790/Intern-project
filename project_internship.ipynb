```python
import dash
from dash import Input, Output, dcc, html
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Heart Disease Dashboard", style={'textAlign': 'center'}),

    # Age Distribution
    dcc.Graph(
        id='age-distribution',
        figure=px.histogram(df, x='age', title='Age Distribution',
                             color='target',
                             color_discrete_sequence=['#636EFA', '#EF553B'],  # Custom colors
                             labels={'age': 'Age'},
                             barmode='overlay')
    ),

    # Cholesterol Distribution
    dcc.Graph(
        id='cholesterol-distribution',
        figure=px.histogram(df, x='chol', title='Cholesterol Distribution',
                             color='target',
                             color_discrete_sequence=['#00CC96', '#AB63FA'],
                             labels={'chol': 'Cholesterol (mg/dl)'},
                             barmode='overlay')
    ),

    # Blood Pressure vs Target
    dcc.Graph(
        id='blood-pressure-vs-target',
        figure=px.box(df, x='trestbps', y='target', title='Blood Pressure vs Heart Disease',
                       color='target',
                       color_discrete_sequence=['#FFA07A', '#20B2AA'],
                       labels={'trestbps': 'Resting Blood Pressure (mm Hg)', 'target': 'Heart Disease'})
    ),

    # Gender Distribution
    dcc.Graph(
        id='gender-distribution',
        figure=px.bar(df, x='sex', color='target', title='Gender Distribution of Heart Disease',
                       color_discrete_sequence=['#FFA07A', '#20B2AA'],
                       labels={'sex': 'Gender', 'target': 'Heart Disease'},
                       category_orders={'sex': ['0', '1']})
    ),

    # Chest Pain Type Distribution
    dcc.Graph(
        id='chest-pain-type',
        figure=px.histogram(df, x='cp', title='Chest Pain Type Distribution',
                             color='target',
                             color_discrete_sequence=px.colors.qualitative.Plotly,
                             labels={'cp': 'Chest Pain Type'},
                             barmode='overlay')
    )
])

if __name__ == '__main__':
    app.run(debug=True)
```
